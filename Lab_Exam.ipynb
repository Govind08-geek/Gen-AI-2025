```python
!pip install --upgrade google-generativeai
model = genai.GenerativeModel('gemini-1.5-flash')
import google.generativeai as genai

import google.generativeai as genai
import os
import subprocess

# Replace "YOUR_API_KEY" with the key you obtained from Google AI Studio
GEMINI_API_KEY = "AIzaSyD5G6sdClAtp8GMsBkEt5gD1LkTtbSpCPQ"

if not GEMINI_API_KEY:
    raise ValueError("Gemini API Key not set. Please replace 'YOUR_API_KEY' with your actual key from Google AI Studio.")

try:
    genai.configure(api_key=GEMINI_API_KEY)
    model = genai.GenerativeModel('gemini-1.5-flash')
except ImportError:
    print("\nError: The 'google-generativeai' library is not installed.")
    print("Please open your terminal or command prompt and run:")
    print("  pip install google-generativeai")
    print("Then, run this script again.")
    exit()
except Exception as e:
    print(f"\nError configuring Gemini API: {e}")
    exit()

def generate_paragraph(prompt):
    """Generates a paragraph using the Gemini API based on the given prompt."""
    try:
        response = model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"Error generating text: {e}"

def analyze_coherence_fluency(text):
    """
    Analyzes the coherence and fluency of the generated text.
    (Same analysis function as before)
    """
    coherence_score = 0
    fluency_score = 0
    analysis = ""

    if not text:
        return 0, 0, "No text to analyze."

    sentences = text.split('.')
    sentences = [s.strip() for s in sentences if s.strip()]

    if len(sentences) < 2:
        analysis = "The paragraph is too short to effectively assess coherence and fluency."
        return 0, 0, analysis

    keywords = text.lower().split()
    unique_keywords = set(keywords)
    keyword_density = len(unique_keywords) / len(keywords) if keywords else 0
    if keyword_density > 0.3:
        coherence_score += 0.4
        analysis += "- The paragraph shows some thematic focus with recurring keywords.\n"
    else:
        analysis += "- The thematic focus could be stronger.\n"

    connectors = ["however", "therefore", "in addition", "furthermore", "consequently", "as a result", "for example", "in conclusion"]
    connector_count = sum(1 for word in keywords if word in connectors)
    if connector_count > 0:
        coherence_score += 0.3
        analysis += "- The presence of logical connectors aids in the flow of ideas.\n"
    else:
        analysis += "- More explicit logical connectors could improve coherence.\n"

    avg_sentence_length = sum(len(s.split()) for s in sentences) / len(sentences) if sentences else 0
    if 10 <= avg_sentence_length <= 25:
        fluency_score += 0.5
        analysis += "- The average sentence length suggests good readability.\n"
    elif avg_sentence_length < 10:
        analysis += "- The sentences might be too short and choppy.\n"
    else:
        analysis += "- Some sentences might be too long and complex.\n"

    first_words = [s.split()[0].lower() for s in sentences if s.split()]
    if len(first_words) == len(set(first_words)):
        fluency_score += 0.3
        analysis += "- There is good variation in sentence beginnings.\n"
    else:
        analysis += "- Repetitive sentence beginnings could impact fluency.\n"

    coherence_score = max(0, min(1, coherence_score))
    fluency_score = max(0, min(1, fluency_score))

    return coherence_score, fluency_score, analysis

if __name__ == "__main__":
    prompt = input("Enter your prompt for text generation: ")
    generated_text = generate_paragraph(prompt)

    print("\nGenerated Paragraph:")
    print(generated_text)
    print("\n--- Analysis ---")

    coherence, fluency, analysis = analyze_coherence_fluency(generated_text)
    print(f"Coherence Score (out of 1): {coherence:.2f}")
    print(f"Fluency Score (out of 1): {fluency:.2f}")
    print("Qualitative Analysis:")
    print(analysis)
```
